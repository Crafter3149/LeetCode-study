```python
import numpy as np
import pandas as pd
import math
```

```python
#load data split data
all_df = pd.read_csv("KS/KS.csv")
all_df.apply(pd.to_numeric, errors='ignore')
ndarray = all_df.values
```

```python
r = []
n2 = np.mean(ndarray[:,19])
for i in range(16):
    c=0
    b=0
    a=0
    n1 = np.mean(ndarray[:,i+3])
    print(i , end = '\r')
    for j in range(len(ndarray)):

        c = c + (ndarray[j,i+3]-n1)*(ndarray[j,19]-n2)
        b = b + (ndarray[j,i+3]-n1)**2
        a = a + (ndarray[j,19] -n2)**2
    r.append(c/(a*b)**0.5)
```

```python
#Save Result
writer = pd.ExcelWriter('r.xlsx')
df = pd.DataFrame(list(all_df.columns.values)[3:19])
df.insert(0,'r',r)
df.insert(0,'up',np.linspace(0.3,0.3,16))
df.insert(0,'down',np.linspace(-0.3,-0.3,16))
df.to_excel(writer,sheet_name='sheet1')
```
